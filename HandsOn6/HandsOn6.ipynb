```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        #picking pivot last element
        pivot = arr[-1]
        smaller_than_pivot = []
        greater_than_pivot = []

        for x in arr[:-1]:
            if x <= pivot:
                smaller_than_pivot.append(x)  
            else:
                greater_than_pivot.append(x)  

        return quicksort(smaller_than_pivot) + [pivot] + quicksort(greater_than_pivot)

arr = [6,9,3,5,2,1,8,7]
res = quicksort(arr)
print("sorted array after choosing non random pivot element: ", res)
```

```python
import random

def quicksort_random(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = random.choice(arr)  
        smaller_or_equal_pivot = []  
        greater_than_pivot = [] 
        
        for x in arr:
            if x <= pivot:  
                smaller_or_equal_pivot.append(x) 
            else:  
                greater_than_pivot.append(x) 
        return quicksort_random(smaller_or_equal_pivot) + quicksort_random(greater_than_pivot)

arr = [6,9,7,2,3,1,8,4]
res_R = quicksort_random(arr)
print("sorted array after choosing random pivot element: ", res_R)
```

```python
import time
import random
import numpy as np
import matplotlib.pyplot as plt


def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        #picking pivot last element
        pivot = arr[-1]
        smaller_than_pivot = []
        greater_than_pivot = []

        for x in arr[:-1]:
            if x <= pivot:
                smaller_than_pivot.append(x)  
            else:
                greater_than_pivot.append(x)  

        return quicksort(smaller_than_pivot) + [pivot] + quicksort(greater_than_pivot)

input_sizes = np.arange(100, 4001, 300)  
best_time = []
worst_time = []
average_time = []

# Benchmarking
for n in input_sizes:
    # Best case 
    input1 = list(range(n))
    time_start = time.time()
    quicksort(input1)
    best_time.append(time.time() - time_start)

    # Worst case 
    input2 = list(range(n, 0, -1))
    time_start = time.time()
    quicksort(input2)
    worst_time.append(time.time() - time_start)

    # Average case 
    input3 = list(range(n))
    random.shuffle(input3)
    time_start = time.time()
    quicksort(input3)
    average_time.append(time.time() - time_start)

plt.plot(input_sizes, best_time, label="Best Case")
plt.plot(input_sizes, worst_time, label="Worst Case")
plt.plot(input_sizes, average_time, label="Average Case")
plt.legend()
plt.xlabel('Input Size (n)')
plt.ylabel('Time (in seconds)')
plt.title('Quick Sort for non random pivot')
plt.show()
```
